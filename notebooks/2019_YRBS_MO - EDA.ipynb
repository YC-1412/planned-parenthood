```python
import pandas as pd
```

```python
data = pd.read_csv("../data/2019_YRBS_MO.csv")
```

```python
data.head()
```

```python
data = data.filter(["q1", "q2", "q3", "q19", "q22", "q58", "q59", "q61", 
                    "q62", "q63", "q64", "q65",
                    "q66", "q84", "q92", "raceeth"])
```

```python
data.head()
```

```python
data = data.rename(columns = {"q1": "age", "q2": "gender", 
            "q3": "grade", 
            "q19": "physically_forced_sexual_intercourse",
            "q22": "dating_physically_hurt", 
            "q58": "sexual_intercourse_binary",
            "q59": "sexual_intercourse_age",
            "q61": "sexual_intercourse_number",
            "q62": "sexual_intercourse_alcohol/drugs",
            "q63": "sexual_intercourse_condom",
            "q64": "pregnancy_prevention_method",
            "q65": "sexual_contact_gender",
            "q66": "sexuality",
            "q84": "tested_HIV",
            "q92": "exchange_sex"})
```

```python
data.head()
```

```python
data["age"] = data["age"].replace({1:"12 or younger", 2:"13", 3:"14", 4:"15",
                    5:"16", 6:"17", 7:"18 or older"})
```

```python
data["gender"] = data["gender"].replace({1:"Female", 2:"Male"})
```

```python
data["grade"] = data["grade"].replace({1:"9th", 2:"10th",
                                      3:"11th", 4:"12th",
                                      5:"Ungraded or other"})
```

```python
data["physically_forced_sexual_intercourse"] = data["physically_forced_sexual_intercourse"].replace({1:"Yes", 2:"No"})
```

```python
data["dating_physically_hurt"] = data["dating_physically_hurt"].replace({1:"Did not date or go out in past 12 months", 
                                                                        2:"0 times",
                                                                        3:"1 time",
                                                                        4:"2 or 3 times",
                                                                        5:"4 or 5 times",
                                                                        6:"6 or more times"})
```

```python
data["sexual_intercourse_binary"] = data["sexual_intercourse_binary"].replace({1:"Yes", 2:"No"})
```

```python
data["sexual_intercourse_age"] = data["sexual_intercourse_age"].replace({1:"Never had sexual intercourse", 2:"11 or younger", 
                                                                        3:"12", 4:"13", 5:"14", 6:"15",
                                                                        7:"16", 8:"17 or older"})
```

```python
data["sexual_intercourse_number"] = data["sexual_intercourse_number"].replace({1:"Never had sexual intercourse", 2:"Not in last 3 months", 
                                                                              3:"1 person", 4:"2 people", 5:"3 people",
                                                                              6:"4 people", 7:"5 people", 8: "6 or more people"})
```

```python
data["sexual_intercourse_alcohol/drugs"] = data["sexual_intercourse_alcohol/drugs"].replace({1:"Never had sexual intercourse", 2:"Yes", 3:"No"})
```

```python
data["sexual_intercourse_condom"] = data["sexual_intercourse_condom"].replace({1:"Never had sexual intercourse", 2:"Yes", 3:"No"})
```

```python
data["pregnancy_prevention_method"] = data["pregnancy_prevention_method"].replace({1:"Never had sexual intercourse",
                                                                                   2:"No method",
                                                                                   3:"Birth control pills",
                                                                                   4:"Condoms",
                                                                                   5:"IUD or implant",
                                                                                   6:"Shot, patch, or birth control ring",
                                                                                   7:"Withdrawal or other",
                                                                                   8:"Not sure"
                                                                                  })
```

```python
data["sexual_contact_gender"] = data["sexual_contact_gender"].replace({1:"Never had sexual contact",
                                                                        2:"Females",
                                                                        3:"Males",
                                                                        4:"Females and males"})
```

```python
data["sexuality"] = data["sexuality"].replace({1:"Heterosexual",
                                                2:"Gay or lesbian",
                                                3:"Bisexual",
                                                4:"Not sure"})
```

```python
data["tested_HIV"] = data["tested_HIV"].replace({1:"Yes", 2:"No", 3:"Not sure"})
```

```python
data["exchange_sex"] = data["exchange_sex"].replace({1:"Yes", 
                                                 2:"No"})
```

```python
data["raceeth"] = data["raceeth"].replace({1:"American Indian or Alaska Native", 
                                           2:"Asian",
                                           3:"Black or African American",
                                           4:"Native Hawaiian or Other Pacific Islander",
                                           5:"White",
                                           6:"Other",
                                           7:"Hispanic or Latino",
                                           8:"Multiple - Hispanic",
                                           9:"Multiple - non-Hispanic"
                                          })
```
